```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters
N= 8   #طول و عرض
k=1      #ثابت بولتزمن
steps=20000 # Number of MC sweeps 
T=0    #Temperature
u=1  #magnetic moment of each spin

#ثابت های تعامل اسپین
J1=1
J2=0.5
J3=0.2
```

```python
def initialstate(N):        #generating random spins for initial condition
    return np.random.choice([1, -1], size=(N, N))    

def deltaE(S0, Sn, J, H):
    return 2 * S0 * (H + J * Sn)

def ising(N,steps,J1,J2,J3,T):
    f = initialstate(N)
    energy = 0
    energies = []
    spins = []
    spin = np.sum(f)
    for step in range(steps):
        i = np.random.randint(N)
        j = np.random.randint(N)
        # Periodic Boundary Condition
        Sn_list =[J1*f[(i - 1) % N, j], J1*f[(i + 1) % N, j] ,
        J1*f[i, (j - 1) % N] , J1*f[i, (j + 1) % N] +
        J2*f[(i - 1) % N, (j - 1)%n] , J2*f[(i + 1) % N,(j - 1)%N] ,
        J2*f[(i - 1) % N, (j + 1)%n] , J2*f[(i + 1) % N,(j + 1)%N] ,
        J3*f[(i - 2) % N, j] , J3*f[(i + 2) % N, j] ,
        J3*f[i, (j - 2) % N] , J3*f[i, (j + 2) % N]]
        Sn=sum(Sn_list)
        S0=f[i, j]
        dE =  2 * S0 * Sn
        if dE < 0 or np.random.random() < np.exp(-dE/(T*k)):
            f[i, j] = -f[i, j]
            energy += dE
            spin += 2*f[i, j]
        spins.append(spin)
        energies.append(energy)
    return  spins , energies
    
```

```python
#random initial state
state = initialstate(N)
fig, ax = plt.subplots(figsize=(9,7))
ax.matshow(state, cmap = plt.cm.plasma) 
ax.set_xticks([])
ax.set_yticks([])
plt.title(f"A random state with {N}*{N}", fontsize = 20, fontweight = "bold" )
plt.savefig("images.png")
plt.show()
```
